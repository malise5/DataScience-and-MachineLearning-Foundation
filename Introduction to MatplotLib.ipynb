# MatplotLib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot();
```

```python
plt.plot([1,2,3,4]);
```

```python
x = [1,2,3,4]
y = [11,12,33,44]
plt.plot(x, y);
```

```python
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x,y);
```

```python
#plot the data
x = [1,2,3,4]
y = [23,45,78,89]

#setup plot
fig, ax = plt.subplots(figsize=(10,10))

#plot the data
ax.plot(x,y)

#customize
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

#save and show
fig.savefig("sample-plot.png")
```

```python
#create data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2);
#line plot
```

```python
#scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
#Scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
#make a plot from dictionary
butter = {"Almond butter": 10, 
          "Peanut Butter": 8,
          "Cashew Butter": 12
         }
fig, ax = plt.subplots()
ax.bar(butter.keys(), butter.values())
ax.set(title="My Butter Budget", xlabel="Butters", ylabel="Butter Prices($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(butter.keys()), list(butter.values()));
```

```python
#histogram plot
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```
